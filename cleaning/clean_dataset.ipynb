```python
import pandas as pd

df = pd.read_csv("../data/raw/reviews_raw.csv")
```

```python
df
```

```python
df = df.dropna(subset=['Rating'])
df
```

```python
df.loc[166]
```

```python
def count_words(review):
    return len(str(review).split())
```

```python
df = df[df['Review'].apply(count_words) >= 12]
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.hist(df['Rating'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Rating Values')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Word_Count'] = df['Review'].apply(count_words)

plt.figure(figsize=(8, 5))
plt.hist(df['Word_Count'], bins=10, color='lightcoral', edgecolor='black')
plt.title('Distribution of Number of Words in Reviews')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()
```

```python
max_words = df['Word_Count'].argmax()
max_words
```

```python
df = df[df['Word_Count'] <= 500]
df
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Word_Count'], bins=10, color='lightcoral', edgecolor='black')
plt.title('Distribution of Number of Words in Reviews')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')
plt.show()
```

```python
df = df.drop(columns=['Word_Count'])
```
